```python
df = gdp[gdp['Year'] == 2016].iloc[46:]
```

```python
df['Value']
```

```python
digits = df['Value'].apply(lambda x: int(str(x)[0]))
```

```python
for i in range(1,10):
    print(sum(digits == i)/len(digits)*100)
```

```python
import matplotlib.pyplot as plt

digits.hist(bins = 9,density = True)
plt.savefig('observed_distr.png')
plt.show()
```

```python
import matplotlib.pyplot as plt

x = np.linspace(1.5,9, 100)
plt.plot(x, np.log(1+1/(x-1/2))/np.log(10),'r-', lw=1, alpha=0.6)
digits.hist(bins = 9,density = True)
plt.savefig('theoretical_distr.png')
plt.show()

```
